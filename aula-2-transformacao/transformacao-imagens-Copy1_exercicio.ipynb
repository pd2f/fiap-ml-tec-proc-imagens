## Operações lógicas

As operações lógicas ou bitwise operations são operações do tipo and, or, xor e not. São utilizadas na composição de 2 imagens, criar máscaras e intersecções.

```python
imagem = cv2.imread("gize.png")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.imshow(imagem)
plt.title("Piramide")
```

```python

height = imagem.shape[0]
width = imagem.shape[1]
triangulo = np.zeros((height,width), np.uint8)
pontos = np.array([[60,410], [300,150], [520,420]], np.int32)
cv2.fillPoly(triangulo,[pontos],(100, 50, 200))

plt.imshow(triangulo)
plt.title("Triangulo")
```
